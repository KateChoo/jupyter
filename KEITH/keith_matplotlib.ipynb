```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3]
y= [0, 2,4,6]
plt.plot(x, y, label='2x', color='maroon', linewidth=2, marker='.', markersize=10)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

##line number 2
x2 = np.arange(0,4,0.5)
print(x2)
plt.plot(x2[:6], x2[:6]**2,'r', label='X^2')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```
